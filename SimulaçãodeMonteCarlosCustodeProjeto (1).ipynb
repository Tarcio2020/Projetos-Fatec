```python
import numpy as np
import matplotlib.pyplot as plt

# Valores prováveis das atividades
valores_provaveis = [
    6000.0,   # Atividade 1.1
    3750.0,   # Atividade 1.2
    16500.0,  # Atividade 1.3
    225000.0, # Atividade 1.4
    35000.0,  # Atividade 2.1
    12500.0,  # Atividade 2.2
    55000.0,  # Atividade 2.3
    17500.0,  # Atividade 2.4
    475000.0, # Atividade 3.1
    300000.0, # Atividade 3.2
    225000.0, # Atividade 3.3
    800000.0, # Atividade 3.4
    125000.0, # Atividade 4.1
    450000.0, # Atividade 4.2
    725000.0, # Atividade 4.3
    350000.0, # Atividade 5.1
    275000.0, # Atividade 5.2
    1100000.0,# Atividade 5.3
    350000.0, # Atividade 6.1
    1050000.0,# Atividade 6.2
    850000.0, # Atividade 6.3
    425000.0, # Atividade 7.1
    375000.0, # Atividade 7.2
    275000.0, # Atividade 7.3
    750000.0, # Atividade 7.4
    250000.0, # Atividade 7.5
    275000.0, # Atividade 8.1
    225000.0, # Atividade 8.2
    450000.0, # Atividade 8.3
    175000.0, # Atividade 8.4
    550000.0, # Atividade 8.5
    125000.0, # Atividade 9.1
    125000.0, # Atividade 9.2
    125000.0, # Atividade 9.3
]

# Número de simulações
num_simulacoes = 2000

# Lista para armazenar os custos totais de cada simulação
custos_totais = []

# Realizar as simulações
for _ in range(num_simulacoes):
    custo_total_simulacao = 0
    for valor in valores_provaveis:
        custo_simulacao = np.random.uniform(0.7, 1.3) * valor  # Variação de 20% para cima ou para baixo
        custo_total_simulacao += custo_simulacao
    custos_totais.append(custo_total_simulacao)

# Plotar histograma dos custos totais
plt.figure(figsize=(10, 6))
plt.hist(custos_totais, bins=30, edgecolor='black', density=True)
plt.xlabel('Custo Total do Projeto')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição dos Custos Totais do Projeto')
plt.grid(True)
plt.show()

# Plotar gráfico de densidade dos custos totais
plt.figure(figsize=(10, 6))
plt.hist(custos_totais, bins=30, edgecolor='black', density=True, alpha=0.5)
plt.plot(sorted(custos_totais), np.linspace(0, 1, len(custos_totais)), color='red', linewidth=2)
plt.xlabel('Custo Total do Projeto')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição de Densidade dos Custos Totais do Projeto')
plt.grid(True)
plt.show()
```

```python
# Orçamento
orcamento = 11571250.0

# Contar o número de simulações que estão dentro do orçamento
num_simulacoes_dentro_orcamento = sum(custo <= orcamento for custo in custos_totais)

# Calcular a probabilidade de entregar o projeto dentro do orçamento
probabilidade = num_simulacoes_dentro_orcamento / num_simulacoes

print("Probabilidade de entregar o projeto dentro do orçamento:", probabilidade)
```
