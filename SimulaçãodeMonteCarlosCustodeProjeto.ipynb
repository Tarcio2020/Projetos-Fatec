```python
import pandas as pd

# Caminho do arquivo Excel
caminho_arquivo = "/content/Análise_Monte_Carlo_Projeto_Gestão.xlsx"

# Ler a planilha específica do Excel
dados_excel = pd.read_excel(caminho_arquivo)


# Exibir os dados lidos
display(dados_excel)
```

```python
# Excluir linhas onde as colunas 3, 4 e 5 estão vazias
dados_limpos = dados_excel.dropna(subset=[dados_excel.columns[2], dados_excel.columns[3], dados_excel.columns[4]])

# Exibir os dados após a exclusão das linhas

coluna_4 = dados_limpos.iloc[:, 3]  # O índice 3 representa a quarta coluna (começando de 0)

# Exibir a coluna filtrada
print(coluna_4)
```

```python
soma_coluna_4 = dados_limpos[dados_limpos.columns[3]].sum()

# Exibir a soma dos valores da coluna 4
print("Soma dos valores da coluna 4:", soma_coluna_4)

soma_coluna_5 = dados_limpos[dados_limpos.columns[4]].sum()

# Exibir a soma dos valores da coluna 4
print("Soma dos valores da coluna 5:", soma_coluna_5)

soma_coluna_6 = dados_limpos[dados_limpos.columns[5]].sum()

# Exibir a soma dos valores da coluna 4
print("Soma dos valores da coluna 6:", soma_coluna_6)
```

```python
import numpy as np
import pandas as pd

# Definindo o número de simulações
num_simulacoes = 30

# Definindo o índice para a coluna provável
indice_provavel = 4

# Convertendo a coluna provável em lista
valores_provaveis = dados_limpos.iloc[:, indice_provavel].tolist()

# Resultados das simulações
resultados_simulacoes = []

# Loop para realizar as simulações
for _ in range(num_simulacoes):
    # Gerando um valor aleatório para o cenário provável com variação de até 30%
    valor_provavel_simulacao = np.random.uniform(0.7, 1.3) * np.random.choice(valores_provaveis)

    # Adicionando o valor total ao resultado das simulações
    resultados_simulacoes.append(valor_provavel_simulacao)

# Convertendo os resultados das simulações em um DataFrame
df_resultados_simulacoes = pd.DataFrame(resultados_simulacoes, columns=["Valor Total"])

# Exibindo os resultados das simulações
display(df_resultados_simulacoes)

```

```python
# Definir o orçamento do projeto (valor da coluna 5)
orcamento = 11571250.0

# Contar quantas simulações resultaram em valores menores ou iguais ao orçamento
num_simulacoes_dentro_orcamento = (df_resultados_simulacoes <= orcamento).sum().sum()

# Calcular a probabilidade de entregar o projeto dentro do orçamento
probabilidade_entregar_dentro_orcamento = num_simulacoes_dentro_orcamento / (num_simulacoes * len(df_resultados_simulacoes))

# Exibir a probabilidade de entregar o projeto dentro do orçamento
print("Probabilidade de entregar o projeto dentro do orçamento:", probabilidade_entregar_dentro_orcamento)
```
