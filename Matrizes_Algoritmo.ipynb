```python
def ler_matriz(nome):
    """Lê uma matriz 2x2 de inteiros do usuário."""
    matriz = []
    print(f"\nDigite os elementos da matriz {nome}:")
    for i in range(2):
        linha = []
        for j in range(2):
            elemento = int(input(f"Elemento [{i+1}, {j+1}] de {nome}: "))
            linha.append(elemento)
        matriz.append(linha)
    return matriz

def exibir_matriz(matriz, nome):
    """Exibe a matriz de forma formatada."""
    print(f"\nMatriz {nome}:")
    for linha in matriz:
        for elemento in linha:
            print(f"{elemento} ", end="")
        print()  # Nova linha após cada linha da matriz

def somar_matrizes(A, B):
    """Soma duas matrizes 2x2."""
    C = []
    for i in range(2):
        linha = []
        for j in range(2):
            linha.append(A[i][j] + B[i][j])
        C.append(linha)
    return C

# Função principal
def main():
    # Ler as matrizes A e B
    matriz_A = ler_matriz("A")
    matriz_B = ler_matriz("B")

    # Exibir as matrizes A e B
    exibir_matriz(matriz_A, "A")
    exibir_matriz(matriz_B, "B")

    # Somar as matrizes A e B para obter a matriz C
    matriz_C = somar_matrizes(matriz_A, matriz_B)

    # Exibir a matriz C
    exibir_matriz(matriz_C, "C")

if __name__ == "__main__":
    main()
```

```python
def ler_matriz(linhas, colunas, nome):
    """Lê uma matriz de dimensões especificadas do usuário."""
    print(f"\nDigite os elementos da matriz {nome} ({linhas}x{colunas}):")
    matriz = []
    for i in range(linhas):
        linha = []
        for j in range(colunas):
            elemento = int(input(f"Elemento [{i+1}, {j+1}] de {nome}: "))
            linha.append(elemento)
        matriz.append(linha)
    return matriz

def exibir_matriz(matriz, nome):
    """Exibe a matriz de forma formatada."""
    print(f"\nMatriz {nome}:")
    for linha in matriz:
        for elemento in linha:
            print(f"{elemento} ", end="")
        print()

def multiplicar_matrizes(A, B):
    """Multiplica duas matrizes A e B compatíveis."""
    linhas_A = len(A)
    colunas_A = len(A[0])
    linhas_B = len(B)
    colunas_B = len(B[0])

    # Inicializa a matriz C com zeros
    C = [[0 for _ in range(colunas_B)] for _ in range(linhas_A)]

    # Realiza a multiplicação
    for i in range(linhas_A):
        for j in range(colunas_B):
            for k in range(colunas_A):
                C[i][j] += A[i][k] * B[k][j]
    return C

def main():
    # Ler dimensões da matriz A
    linhas_A = int(input("Digite o número de linhas da matriz A: "))
    colunas_A = int(input("Digite o número de colunas da matriz A: "))

    # Ler dimensões da matriz B
    linhas_B = int(input("Digite o número de linhas da matriz B: "))
    colunas_B = int(input("Digite o número de colunas da matriz B: "))

    # Verificar compatibilidade para multiplicação
    if colunas_A != linhas_B:
        print("Erro: As matrizes A e B não são compatíveis para multiplicação.")
        return

    # Ler elementos das matrizes
    matriz_A = ler_matriz(linhas_A, colunas_A, "A")
    matriz_B = ler_matriz(linhas_B, colunas_B, "B")

    # Exibir as matrizes A e B
    exibir_matriz(matriz_A, "A")
    exibir_matriz(matriz_B, "B")

    # Multiplicar as matrizes
    matriz_C = multiplicar_matrizes(matriz_A, matriz_B)

    # Exibir a matriz resultante
    exibir_matriz(matriz_C, "C")

if __name__ == "__main__":
    main()
```
